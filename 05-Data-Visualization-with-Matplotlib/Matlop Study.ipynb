```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)

axes[0].set_title('first x')
axes[1].set_title('first y')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='x to y ')

ax.plot(x**4,y,label='x to the power of 4')
ax.plot(y**2,x, label='y squre')

ax.legend(loc=0)
```
