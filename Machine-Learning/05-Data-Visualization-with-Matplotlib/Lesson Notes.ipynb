```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2;
```

```python
print(x)
print(y)
```

```python
# FUNCTIONAL
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger title')

axes2.plot(y,x)
axes2.set_title('Smaller title')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Title')

axes[1].plot(y,x)
axes[1].set_title('Second Title')
```

Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label= 'X squared')
ax.plot(x,x**3, label= 'X cubed')

ax.legend(loc=0)
```

```python
fig.savefig('my_pictire.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='b', linewidth=3, alpha=0.4, linestyle='--', marker='o', markersize=4,markerfacecolor='yellow',
        markeredgewidth=3)

```
